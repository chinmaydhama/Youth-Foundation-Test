```python
import pandas as pd

df_teacher_submission = pd.read_csv("teacher_submissions.csv")
print(df_teacher_submission.head())

df_teacher_registration= pd.read_csv("teacher_registration (1).csv")
print(df_teacher_registration.head())
```

```python
import json

def extract_corrected_school_info(json_str):
    try:
        json_data = json.loads(json_str.replace("'", '"'))  # Convert to valid JSON
        school_info = {
            'division': None,
            'district': None,
            'block': None,
            'cluster': None,
            'school': None
        }
        for item in json_data:
            if item['label'] == 'teacher_division':
                school_info['division'] = item['value']
            elif item['label'] == 'teacher_district':
                school_info['district'] = item['value']
            elif item['label'] == 'teacher_block':
                school_info['block'] = item['value']
            elif item['label'] == 'teacher_cluster':
                school_info['cluster'] = item['value']
            elif item['label'] == 'teacher_school':
                school_info['school'] = item['value']
        return school_info
    except Exception:
        return {'division': None, 'district': None, 'block': None, 'cluster': None, 'school': None}

corrected_school_details = df_teacher_registration['results'].apply(
    lambda x: extract_corrected_school_info(x) if isinstance(x, str) else {}
)
corrected_school_details_df = pd.DataFrame(corrected_school_details.tolist())
df_teacher_registration = pd.concat([df_teacher_registration, corrected_school_details_df], axis=1)
df_teacher_registration.drop(columns=['results'], inplace=True)
df_teacher_registration.head()
```

```python
df_teacher_registration
```

```python
duplicate_teachers = df_teacher_registration[df_teacher_registration.duplicated(subset=['id'], keep=False)]

duplicate_teachers
```

```python

df_teacher_registration['inserted_at'] = pd.to_datetime(df_teacher_registration['inserted_at'])
df_teacher_registration = df_teacher_registration.sort_values(by='inserted_at', ascending=False)
df_teacher_registration = df_teacher_registration.drop_duplicates(subset=['id'], keep='first')
```

```python
df_merged = df_teacher_submission.merge(
    df_teacher_registration[['id', 'division', 'district', 'block', 'cluster', 'school']],
    left_on='contact_id',
    right_on='id',
    how='left'
)
```

```python
df_merged
```

```python
df_merged.drop(columns=['id'], inplace=True)
```

```python
df_merged
```

```python

def extract_assessment_info(json_str):
    try:
        json_data = json.loads(json_str)
        assessment_info = {
            'baseline_level': json_data.get('bl_student_level', {}).get('category', None),
            'tutoring_confirmed': json_data.get('impl_lt_confirm', {}).get('category', None),
            'endline_level': json_data.get('el_student_level', {}).get('category', None)
        }
        return assessment_info
    except Exception:
        return {'baseline_level': None, 'tutoring_confirmed': None, 'endline_level': None}
    
df_merged[['baseline_level', 'tutoring_confirmed', 'endline_level']] = df_merged['results'].apply(
    lambda x: pd.Series(extract_assessment_info(x) if isinstance(x, str) else {})
)
df_merged.drop(columns=['results'], inplace=True)


```

```python
df_merged
```

```python
df_merged.columns
```

```python

level_mapping = {
    'Beginner': 0,
    'Addition': 1,
    'Subtraction': 2,
    'Multiplication': 3,
    'Division': 4
}
df_merged['baseline_level_numeric'] = df_merged['baseline_level'].map(level_mapping)
df_merged['endline_level_numeric'] = df_merged['endline_level'].map(level_mapping)
```

```python
# Calculate the percentage of rows with missing school information
missing_schools = df_merged['school'].isnull().sum()
total_rows = len(df_merged)
missing_percentage = (missing_schools / total_rows) * 100
missing_percentage
```

```python
df_merged = df_merged.dropna(subset=['school'])
```

```python
df_merged
```

```python
df_merged['week'] = pd.to_datetime(df_merged['inserted_at']).dt.to_period('W')
df_final = df_merged.groupby(['school', 'week']).agg(
    teachers_registered=('contact_id', 'nunique'), 
    baseline_assessments=('baseline_level_numeric', 'count'),  
    baseline_average_level=('baseline_level_numeric', 'mean'), 
    tutoring_calls=('tutoring_confirmed', lambda x: (x == 'Yes').sum()),  # Count tutoring calls confirmed
    endline_assessments=('endline_level_numeric', 'count'),
    endline_average_level=('endline_level_numeric', 'mean') 
).reset_index()
```
